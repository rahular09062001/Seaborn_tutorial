# Barplot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

### Available data set 

```python
#sn.get_dataset_names()
test=sn.load_dataset('brain_networks')
test.head()
```

### Loading own data set 

```python
df=pd.read_csv('../used_cars_data.csv')
df.head()
```

### Basic bar ploting

```python
plt.figure(figsize=(15,5))
sn.barplot(x='Year',y='Kilometers_Driven',data=df)
```

### Adding hue

```python
plt.figure(figsize=(15,5))
sn.barplot(x='Year',y='Kilometers_Driven',hue='Transmission',data=df)
```

### Horizontal Plot

```python
plt.figure(figsize=(15,5))
sn.barplot(x='Seats',y='Price', hue='Transmission',data=df)
```

### Plot the bars in given order

```python
df.columns
```

```python
x=df['Location'].sort_values()
x.unique()
```

```python
plt.figure(figsize=(15,5))
sn.barplot(x='Location',y='Kilometers_Driven',data=df,order=['Ahmedabad', 'Bangalore', 'Chennai', 'Coimbatore', 'Delhi',
       'Hyderabad', 'Jaipur', 'Kochi', 'Kolkata', 'Mumbai', 'Pune'])
```

## Changing Hue Order

```python
x=df['Transmission'].sort_values()
x.unique()
```

```python
plt.figure(figsize=(15,5))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',hue_order=['Manual','Automatic'],data=df,order=['Ahmedabad', 'Bangalore', 'Chennai', 'Coimbatore', 'Delhi',
       'Hyderabad', 'Jaipur', 'Kochi', 'Kolkata', 'Mumbai', 'Pune'])
```

## Adding caps on the error bar

```python
df.columns
```

```python
plt.figure(figsize=(15,6))
sn.barplot(x='Location',y='Kilometers_Driven',hue="Transmission",capsize=0.1,data=df)
```

## Removing the Error Bar

errorbar=None

```python
plt.figure(figsize=(15,6))
sn.barplot(x='Location',y='Kilometers_Driven',hue="Transmission",data=df,errorbar=None)
```

## Change the color of the columns 

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',data=df,color='green')
plt.show()
```

## Change the color using palette 

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',data=df,palette='cividis')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',data=df,palette='cividis')
plt.show()
```

## Using saturation parameter to make changes in the saturation of color

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',data=df,palette='cividis',saturation=0.1)
plt.show()
```

## Change the default aggregation method using estimator attribute

By default the estimator is mean \
We can change the estimator to sum, median and so on as we wish

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',data=df,palette='cividis',estimator=np.median)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',data=df,palette='cividis',estimator=sum)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sn.barplot(x='Location',y='Kilometers_Driven',hue='Transmission',data=df,palette='cividis',estimator=np.max,errorbar=None)
plt.show()
```
