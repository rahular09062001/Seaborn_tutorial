# Coutplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

### Importing data

```python
df=pd.read_csv('../used_cars_data.csv')
```

```python
df.head()
```

```python
df.columns
```

### Creating basic count plot 

```python

sn.countplot(x='Year',data=df)
```

```python
plt.figure(figsize=(13,6))
sn.countplot(x='Year',data=df)
```

```python
plt.figure(figsize=(15,8))
sn.countplot(x='Location',data=df)
```

### Horizontal Plot

```python
sn.countplot(y='Location',data=df)
```

# Adding hue to the plot

```python
plt.figure(figsize=(17,5))
sn.countplot(x='Year',hue='Transmission',data=df)
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',hue='Location',data=df)
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',hue='Fuel_Type',data=df)
```

### Different Color Palette

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',hue='Transmission',data=df,palette='pastel')
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',hue='Transmission',data=df,palette='crest')
```

### Change the style using facecolor, linewidth and edge color

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',data=df)
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',data=df,facecolor=(1,0,1,0),linewidth=3,edgecolor=sn.color_palette('dark',3))
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',data=df,facecolor=(1,0,1,0),linewidth=6,edgecolor=sn.color_palette('dark',10))
```

```python
plt.figure(figsize=(15,5))
sn.countplot(x='Year',data=df,linewidth=5,edgecolor=sn.color_palette('dark',10))
```
