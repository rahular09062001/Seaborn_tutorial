# Box plot

This provides a summary of supplied data, which includes the information like:
1. Mean 
  2. Median 
  3. Minimum 
  4. Maximum 
  5. 1st Quartile 
  6. 3rd Quartile 
  7. Outliers

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading data set

```python
df=pd.read_csv('../used_cars_data.csv')
df.head(10)
```

## Basic box plot on one numeric variabel

```python
sns.boxplot(x='Price',data=df,width=0.5)
plt.show()
```

```python
sns.boxplot(y='Price',data=df,width=0.25)
plt.show()
```

## Create a basic box plot on one numberical variable by a categorical variable

```python
sns.boxplot(x='Transmission',y='Price',data=df)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Location',y='Price',data=df)
```

## Create a basic box plot on one numberical variable by two categorical variable using hue

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Location',y='Price',hue='Transmission',data=df)
plt.show()
```

## Add mean marker to the box plot using showmeans attribute and change it's style using meangroups

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Location',y='Price',hue='Transmission',data=df,showmeans=True)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Location',y='Price',hue='Transmission',data=df,showmeans=True,meanprops={"marker":"o",
                                                                                       "markerfacecolor":"white",
                                                                                       "markersize":10,
                                                                                       "markeredgecolor":"black"})
```

## Horizontal boxplots

```python
plt.figure(figsize=(10,15))
sns.boxplot(y='Location',x='Price',hue='Transmission',data=df,showmeans=True,meanprops={'marker':"o",
                                                                                       'markerfacecolor':'white',
                                                                                       'markersize':10,
                                                                                       'markeredgecolor':'black'})
plt.show()
```

## Changing the color palette and line width

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='Location',y='Price',hue='Transmission',data=df,showmeans=True,meanprops={'marker':"o",
                                                                                       'markerfacecolor':'white',
                                                                                       'markersize':10,
                                                                                       'markeredgecolor':'black'},
           palette='muted',
           linewidth=3)
plt.show()
```

## Creating box plot for each of the numeric variable in the dataframe

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df)
plt.show()
```
